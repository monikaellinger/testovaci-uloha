```python
df_scaled_values = df_scaled.values
clustering_scores_stand = []
for k in range(2, 11):
    clustering_stand = sklearn.cluster.KMeans(n_clusters=k, random_state=13).fit(df_scaled_values)
    clustering_scores_stand.append({
        'k': k,
        'sse': clustering_stand.inertia_,
        'silhouette': sklearn.metrics.silhouette_score(df_scaled_values, clustering_stand.labels_)
    })
df_clustering_scores_stand = pd.DataFrame.from_dict(clustering_scores_stand, orient='columns')
df_clustering_scores_stand = df_clustering_scores_stand.set_index('k')
```

```python
sns.lineplot(data = df_clustering_scores_stand, y='sse', x='k')
plt.title('SSE for Standardized Data')
plt.show()
sns.lineplot(data = df_clustering_scores_stand, y='silhouette', x='k')
plt.title('Silhouette Score for Standardized Data')
plt.show()
```

3 clustery

```python
clustering_stand = sklearn.cluster.KMeans(n_clusters=3, random_state=13)
clustering_stand.fit(df_scaled_values)
```

```python
df_scaled['cluster_id'] = clustering_stand.labels_
```

```python
sns.countplot(data=df_scaled, x='cluster_id', hue='child_mort')
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

mean_inflation_per_cluster = df_scaled.groupby('cluster_id')['inflation'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=mean_inflation_per_cluster.index, y=mean_inflation_per_cluster.values)
plt.xlabel('Cluster ID')
plt.ylabel('Inflation')
plt.title('Inflation per Cluster')
plt.show()
```

```python
income_per_cluster = df_scaled.groupby('cluster_id')['income'].median()


plt.figure(figsize=(8, 6))
sns.barplot(x=income_per_cluster.index, y=income_per_cluster.values)
plt.xlabel('Cluster ID')
plt.ylabel('Median Income')
plt.title('Median Income per Cluster')
plt.show()
```
